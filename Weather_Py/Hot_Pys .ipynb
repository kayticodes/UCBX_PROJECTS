```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
## Generate Cities List
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#Perform API Calls 
query_url = "http://api.openweathermap.org/data/2.5/weather?appid="+api_key+"&unitis=imperial"+"&q="

#Create containers to hold data
date = []
citylist=[]
country_code =[]
lat =[]
long =[]
temp_max=[]
humid=[]
cloud= []
wind =[]

#Include a print log of each city as it's being processed with the city number and city name
print("-"*50)
print("Gathering Data")
print ("-"*50)
#Create a loop to go through all the cities in the API request 
counter =0 
for city in cities:
    counter += 1
#print a log of each API request 
    print (f"processing {counter} of {len(cities)} for the city {city}")
    try: 
        response= requests.get(query_url + city).json()
        #Append our data to the list we made earlier 
        citylist.append(response["name"])
        country_code.append(response["sys"]["country"])
        lat.append(response["coord"]["lat"])
        long.append(response["coord"]["lon"])
        temp_max.append(response["main"]["temp_max"])
        humid.append(response["clouds"]["all"])
        wind.append(response["wind"]["speed"])
        date.append(response["dt"])
        cloud.append(response["clouds"]["all"])
    except: 
        print(f"City not found. Skipping...")
print("-"*50)
print("Data Gathering Complete")
print ("-"*50)
```

```python
#Check the number of cities that made it out
len(citylist)
```

```python
#Create a df to add all the data from the API calls

weather_df = pd.DataFrame({"Country Code": country_code,
                          "Date" : date,
                          "City": citylist,
                          "Latitude": lat, 
                          "Longitude":long,
                          "Max Temp": temp_max,
                          "Humidity": humid, 
                          "Cloud Cover": cloud,
                          "Wind Speed": wind})

weather_df.to_csv("Weather_Data.csv", index=False)
weather_df.head()                    
```
